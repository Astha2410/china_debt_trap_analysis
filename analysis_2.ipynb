```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import wordcloud
```

```python
sector_totals = df.groupby('SECTOR')['AMOUNT'].sum()
plt.bar(sector_totals.index, sector_totals.values)
plt.xlabel('Sector')
plt.ylabel('Total Amount')
plt.title('Total Amount Borrowed or Lent by Sector')
plt.show()
```

```python
data = df

text = ' '.join(data['Country'].astype(str).str.lower())

wordcloud = WordCloud(width = 500, height = 200, 
                background_color ='white', 
                stopwords = set(), 
                min_font_size = 10).generate(text)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
